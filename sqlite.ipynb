```python
import sqlite3
```

```python
# create the connection object, if the db doesn't exist it will be created
conn = sqlite3.connect('example.db')
```

```python
# create the cursor object, used for building queries
cur = conn.cursor()
```

```python
cur.execute('CREATE TABLE IF NOT EXISTS test(column1 int, column2 varchar)')
```

```python
conn.commit()
```

```python
cur.execute('INSERT INTO test values(1, "asdf")')
```

```python
conn.commit()
```

```python
cur.execute('SELECT * FROM test')
```

```python
cur.fetchall()
```

```python
conn.close()
```
